# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x = plt.linspace(-10, 10, 100)
```

```python
plt.plot(x, relu(x))
plt.show()
```

```python
def drelu(x):
    return (1 * (x > 0))
```

```python
plt.plot(x, drelu(x))
plt.show()
```
